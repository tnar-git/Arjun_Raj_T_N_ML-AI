```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
csv_file='data.csv'
df=pd.read_csv(csv_file)
df
```

```python
## - format dates

df['Date']=pd.to_datetime(df['Date'])
df
```

```python
## fillna

df['Calories'].fillna(df['Calories'].mean(),inplace=True)
df
```

```python
## correct typos 
row=np.argwhere(df.values == 450)[0][0]
col=np.argwhere(df.values == 450)[0][1]
df.iloc[row,col]=45
df
                                       
```

```python
### finally, remove NaN rows
df.dropna(inplace=True)
df
```

```python
### data visualisation
plt.close()
sns.relplot(data=df,x='Date',y='Calories',hue='Pulse',style='Duration',size='Maxpulse',aspect=2.0)
plt.xticks(fontsize=10)
plt.show()
```
